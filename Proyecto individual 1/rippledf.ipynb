```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

df_ripple = pd.read_json("ripple max.json")


df_ripple["prices"] = df_ripple["prices"].astype(str)
df_ripple["market_caps"] = df_ripple["market_caps"].astype(str)
df_ripple["total_volumes"] = df_ripple["total_volumes"].astype(str)

df_ripple["unix_timestamp"] = df_ripple["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_ripple["date"] = pd.to_datetime(df_ripple["unix_timestamp"], unit="ms")
df_ripple.head(10)

df_ripple["prices"] = df_ripple["prices"].apply(ast.literal_eval)
df_ripple["market_caps"] = df_ripple["market_caps"].apply(ast.literal_eval)
df_ripple["total_volumes"] = df_ripple["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_ripple["prices"] = df_ripple["prices"].apply(remove_unix_timestamp)
df_ripple["market_caps"] = df_ripple["market_caps"].apply(remove_unix_timestamp)
df_ripple["total_volumes"] = df_ripple["total_volumes"].apply(remove_unix_timestamp)

df_ripple = df_ripple.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
pd.options.display.float_format = '{:.15f}'.format
df_ripple['year'] = df_ripple['date'].dt.year
df_ripple.head(10)
```

```python
archivo_csv = "Ripple_DF.csv"
df_ripple.to_csv(archivo_csv, index=False)
```
