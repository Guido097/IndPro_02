```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

df_tether = pd.read_json("tether max.json")


df_tether["prices"] = df_tether["prices"].astype(str)
df_tether["market_caps"] = df_tether["market_caps"].astype(str)
df_tether["total_volumes"] = df_tether["total_volumes"].astype(str)

df_tether["unix_timestamp"] = df_tether["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_tether["date"] = pd.to_datetime(df_tether["unix_timestamp"], unit="ms")
df_tether.head(10)

df_tether["prices"] = df_tether["prices"].apply(ast.literal_eval)
df_tether["market_caps"] = df_tether["market_caps"].apply(ast.literal_eval)
df_tether["total_volumes"] = df_tether["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_tether["prices"] = df_tether["prices"].apply(remove_unix_timestamp)
df_tether["market_caps"] = df_tether["market_caps"].apply(remove_unix_timestamp)
df_tether["total_volumes"] = df_tether["total_volumes"].apply(remove_unix_timestamp)

df_tether = df_tether.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
pd.options.display.float_format = '{:.15f}'.format
df_tether['year'] = df_tether['date'].dt.year
df_tether.head(10)
```

```python
archivo_csv = "Tether_DF.csv"
df_tether.to_csv(archivo_csv, index=False)
```
