```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

with open ("cardano max.json") as f:
    datacardano = json.load(f)


df_cardano = pd.read_json('cardano max.json')

df_cardano.head(10)
```

```python
df_cardano["prices"] = df_cardano["prices"].astype(str)
df_cardano["market_caps"] = df_cardano["market_caps"].astype(str)
df_cardano["total_volumes"] = df_cardano["total_volumes"].astype(str)

df_cardano["unix_timestamp"] = df_cardano["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_cardano["date"] = pd.to_datetime(df_cardano["unix_timestamp"], unit="ms")
df_cardano.head(10)

df_cardano["prices"] = df_cardano["prices"].apply(ast.literal_eval)
df_cardano["market_caps"] = df_cardano["market_caps"].apply(ast.literal_eval)
df_cardano["total_volumes"] = df_cardano["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_cardano["prices"] = df_cardano["prices"].apply(remove_unix_timestamp)
df_cardano["market_caps"] = df_cardano["market_caps"].apply(remove_unix_timestamp)
df_cardano["total_volumes"] = df_cardano["total_volumes"].apply(remove_unix_timestamp)

df_cardano = df_cardano.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
df_cardano['year'] = df_cardano['date'].dt.year
pd.options.display.float_format = '{:.15f}'.format

df_cardano.head(10)
```

```python
archivo_csv = "Cardano_DF.csv"
df_cardano.to_csv(archivo_csv, index=False)
```
