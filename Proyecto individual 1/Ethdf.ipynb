```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

df_eth = pd.read_json("eth max.json")


df_eth["prices"] = df_eth["prices"].astype(str)
df_eth["market_caps"] = df_eth["market_caps"].astype(str)
df_eth["total_volumes"] = df_eth["total_volumes"].astype(str)

df_eth["unix_timestamp"] = df_eth["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_eth["date"] = pd.to_datetime(df_eth["unix_timestamp"], unit="ms")
df_eth.head(10)

df_eth["prices"] = df_eth["prices"].apply(ast.literal_eval)
df_eth["market_caps"] = df_eth["market_caps"].apply(ast.literal_eval)
df_eth["total_volumes"] = df_eth["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_eth["prices"] = df_eth["prices"].apply(remove_unix_timestamp)
df_eth["market_caps"] = df_eth["market_caps"].apply(remove_unix_timestamp)
df_eth["total_volumes"] = df_eth["total_volumes"].apply(remove_unix_timestamp)

df_eth = df_eth.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
pd.options.display.float_format = '{:.15f}'.format
df_eth['year'] = df_eth['date'].dt.year
df_eth.head(10)
```

```python
archivo_csv = "Ethereum_DF.csv"
df_eth.to_csv(archivo_csv, index=False)
```
