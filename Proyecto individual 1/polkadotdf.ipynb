```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

df_polka = pd.read_json("polkadot max.json")


df_polka["prices"] = df_polka["prices"].astype(str)
df_polka["market_caps"] = df_polka["market_caps"].astype(str)
df_polka["total_volumes"] = df_polka["total_volumes"].astype(str)

df_polka["unix_timestamp"] = df_polka["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_polka["date"] = pd.to_datetime(df_polka["unix_timestamp"], unit="ms")
df_polka.head(10)

df_polka["prices"] = df_polka["prices"].apply(ast.literal_eval)
df_polka["market_caps"] = df_polka["market_caps"].apply(ast.literal_eval)
df_polka["total_volumes"] = df_polka["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_polka["prices"] = df_polka["prices"].apply(remove_unix_timestamp)
df_polka["market_caps"] = df_polka["market_caps"].apply(remove_unix_timestamp)
df_polka["total_volumes"] = df_polka["total_volumes"].apply(remove_unix_timestamp)

df_polka = df_polka.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
pd.options.display.float_format = '{:.15f}'.format
df_polka['year'] = df_polka['date'].dt.year
df_polka.head(10)
```

```python
archivo_csv = "Polkadot_DF.csv"
df_polka.to_csv(archivo_csv, index=False)
```
