```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

df_usdc = pd.read_json("usd-coin max.json")


df_usdc["prices"] = df_usdc["prices"].astype(str)
df_usdc["market_caps"] = df_usdc["market_caps"].astype(str)
df_usdc["total_volumes"] = df_usdc["total_volumes"].astype(str)

df_usdc["unix_timestamp"] = df_usdc["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_usdc["date"] = pd.to_datetime(df_usdc["unix_timestamp"], unit="ms")
df_usdc.head(10)

df_usdc["prices"] = df_usdc["prices"].apply(ast.literal_eval)
df_usdc["market_caps"] = df_usdc["market_caps"].apply(ast.literal_eval)
df_usdc["total_volumes"] = df_usdc["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_usdc["prices"] = df_usdc["prices"].apply(remove_unix_timestamp)
df_usdc["market_caps"] = df_usdc["market_caps"].apply(remove_unix_timestamp)
df_usdc["total_volumes"] = df_usdc["total_volumes"].apply(remove_unix_timestamp)

df_usdc = df_usdc.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
pd.options.display.float_format = '{:.15f}'.format
df_usdc['year'] = df_usdc['date'].dt.year
df_usdc.head(10)
```

```python
archivo_csv = "USDCoin_DF.csv"
df_usdc.to_csv(archivo_csv, index=False)
```
