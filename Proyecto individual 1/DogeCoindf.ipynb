```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

df_doge = pd.read_json("DogeCoin.json")


df_doge["prices"] = df_doge["prices"].astype(str)
df_doge["market_caps"] = df_doge["market_caps"].astype(str)
df_doge["total_volumes"] = df_doge["total_volumes"].astype(str)

df_doge["unix_timestamp"] = df_doge["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_doge["date"] = pd.to_datetime(df_doge["unix_timestamp"], unit="ms")
df_doge.head(10)

df_doge["prices"] = df_doge["prices"].apply(ast.literal_eval)
df_doge["market_caps"] = df_doge["market_caps"].apply(ast.literal_eval)
df_doge["total_volumes"] = df_doge["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_doge["prices"] = df_doge["prices"].apply(remove_unix_timestamp)
df_doge["market_caps"] = df_doge["market_caps"].apply(remove_unix_timestamp)
df_doge["total_volumes"] = df_doge["total_volumes"].apply(remove_unix_timestamp)

df_doge = df_doge.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
df_doge['year'] = df_doge['date'].dt.year
pd.options.display.float_format = '{:.15f}'.format
df_doge.head(10)
```

```python
archivo_csv = "DogeCoin_DF.csv"
df_doge.to_csv(archivo_csv, index=False)
```
