```python
import pandas as pd
import numpy as np
import requests
import json
import ast
from pycoingecko import CoinGeckoAPI

df_solana = pd.read_json("solana max.json")


df_solana["prices"] = df_solana["prices"].astype(str)
df_solana["market_caps"] = df_solana["market_caps"].astype(str)
df_solana["total_volumes"] = df_solana["total_volumes"].astype(str)

df_solana["unix_timestamp"] = df_solana["prices"].apply(lambda x: int(x.split(", ")[0][1:]))
df_solana["date"] = pd.to_datetime(df_solana["unix_timestamp"], unit="ms")
df_solana.head(10)

df_solana["prices"] = df_solana["prices"].apply(ast.literal_eval)
df_solana["market_caps"] = df_solana["market_caps"].apply(ast.literal_eval)
df_solana["total_volumes"] = df_solana["total_volumes"].apply(ast.literal_eval)

def remove_unix_timestamp(entry):
    return entry[1]

df_solana["prices"] = df_solana["prices"].apply(remove_unix_timestamp)
df_solana["market_caps"] = df_solana["market_caps"].apply(remove_unix_timestamp)
df_solana["total_volumes"] = df_solana["total_volumes"].apply(remove_unix_timestamp)

df_solana = df_solana.drop(columns=['unix_timestamp'])

pd.set_option('display.float_format', '{:.2f}'.format)
pd.options.display.float_format = '{:.15f}'.format
df_solana['year'] = df_solana['date'].dt.year
df_solana.head(10)
```

```python
archivo_csv = "Solana_DF.csv"
df_solana.to_csv(archivo_csv, index=False)
```
